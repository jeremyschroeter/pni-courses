```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rewards = [0] * 19
rewards.append(1)
rewards = np.array(rewards)
```

$$\Delta V = R_{t+1} + V_{t+1} - V_t$$
$$V_t = V_t + \alpha * \Delta V$$

```python
Vs = np.zeros(20)
alpha = 0.5
```

```python
for i in range(20):
    for t in range(19):
        delta_v = rewards[t + 1] + Vs[t + 1] - Vs[t]
        Vs[t] += alpha * delta_v

    plt.plot(Vs) 
```
