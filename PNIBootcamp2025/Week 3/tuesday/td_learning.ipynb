```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rewards = np.random.choice([0, 1], 20, p=[0.25, 0.75])


def update_v(rewards, alpha=0.5):
    V = np.zeros_like(rewards, dtype=float)
    T = 10

    for t in range(1, len(rewards)):
        delta_v = rewards[t] + V[t] - V[t - 1]
        V[t] = V[t - 1] + alpha * delta_v
    return V



plt.plot(update_v(rewards))
plt.scatter(np.arange(len(rewards)), rewards, c=['g' if r == 1 else 'r' for r in rewards])
```
